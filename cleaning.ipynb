#### Change #1:`rename_columns()` will rename the columns of a dataframe to the values at row 25 but doesn't delete row 25.


```python
def rename_columns(df):
    df.columns = df.iloc[25]
    return df
```

```python
standard_df_change1 = rename_columns(standard_df)
defensive_df_change1 = rename_columns(defensive_df)
goalkeeping_df_change1 = rename_columns(goalkeeping_df)
passing_df_change1 = rename_columns(passing_df)

standard_df_change1.head(30)
```

```python
standard_df_change1.columns
standard_df.head(60)
```

`delete_row()` will delete a row from a dataframe where the Rk column is "Rk"

```python

# Delete row where column Rk is "Rk"
def delete_row(df):
    df = df[df['Rk'] != 'Rk']
    return df

```

```python
standard_df_change2 = delete_row(standard_df_change1)
defensive_df_change2 = delete_row(defensive_df_change1)
goalkeeping_df_change2 = delete_row(goalkeeping_df_change1)
passing_df_change2 = delete_row(passing_df_change1)

standard_df_change2.head(60)
```

`delete_column()` will take the parameters `df`, `column_name`

```python
def delete_column(df, column_name):
    df = df.drop(columns=[column_name])
    return df

```

```python
standard_df_change3 = delete_column(standard_df_change2, 'Rk')
defensive_df_change3 = delete_column(defensive_df_change2, 'Rk')
goalkeeping_df_change3 = delete_column(goalkeeping_df_change2, 'Rk')
passing_df_change3 = delete_column(passing_df_change2, 'Rk')
```
